```python
# Converting data into long format by time point
clean_df.head()
clean_df_long = clean_df

value_cols = [
    col for col in clean_df.columns
    if pd.Series(col).str.contains(r'^[A-Za-z]+\d+_\d+$').any()
]

base_vars = set(
    pd.Series(value_cols)
    .str.extract(r'([A-Za-z]+)\d+_\d+')[0]
)

# Keep only true ID columns
id_cols = [
    col for col in clean_df.columns
    if col not in value_cols and col not in base_vars
]

# Melt to long format
df_long = clean_df.melt(
    id_vars=id_cols,
    value_vars=value_cols,
    var_name="variable",
    value_name="value"
)

# Parse variable names
df_long[["measure", "time", "round_type"]] = (
    df_long["variable"]
    .str.extract(r'([A-Za-z]+)(\d+)_(\d+)')
)

df_long["time"] = df_long["time"].astype(int)
df_long["round_type"] = df_long["round_type"].astype(int)

# Pivot back out
df_long = (
    df_long
    .pivot_table(
        index=id_cols + ["time", "round_type"],
        columns="measure",
        values="value",
        aggfunc="first"
    )
    .reset_index()
)

# Clean column names
df_long.columns.name = None

df_long.head()
```
