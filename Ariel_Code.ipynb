```python
## Ariel's Notebook
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('full_data_speed_dating.csv')
'''
for i in df.columns:
    print(i)
'''

pm_df = pd.DataFrame(columns = ['column', 'percent_missing'])

for i in df.columns:
    percent_missing = sum(pd.isnull(df[i]) == True) / len(df[i])
    new_row = pd.DataFrame({'column': [i], 'percent_missing': [percent_missing]})
    pm_df = pd.concat([pm_df, new_row])

# pm_df.to_csv('per_missing.csv', index = False)
```

```python
print(pm_df['percent_missing'].max())
print(pm_df['percent_missing'].min())
sns.histplot(pm_df['percent_missing'])
```
