```python
# Checking missingness of data in clean_df
clean_pm_df = pd.DataFrame(columns = ['column', 'percent_missing'])

for i in clean_df.columns:
    percent_missing = sum(pd.isnull(clean_df[i]) == True) / len(clean_df[i])
    new_row = pd.DataFrame({'column': [i], 'percent_missing': [percent_missing]})
    clean_pm_df = pd.concat([clean_pm_df, new_row])

print(clean_pm_df['percent_missing'].max())
print(clean_pm_df['percent_missing'].min())

n_over20_missing = 0
for i in clean_pm_df['percent_missing']:
    if i > 0.2:
        n_over20_missing += 1

print('The percentage of columns with > 20% missing data: ' + str(round((n_over20_missing / len(clean_pm_df['percent_missing'])) * 100, 2)))

sns.histplot(clean_pm_df['percent_missing'])
```
